# Pascal’s Triangle — Short Note

## Key Rules
- The first and last element of every row is always `1`.
- Any inner element is formed by the sum of the two elements directly above it:  
  **row[i][j] = row[i-1][j-1] + row[i-1][j]**
- Each row has size **i + 1** (0-indexed).
- Build the triangle row-by-row using the previous row.

## Complexity
- **Time:** O(n²) — every row grows, total cells ≈ n(n+1)/2  
- **Space:** O(n²) — unavoidable since the full triangle is returned.

## Core Concept
Each row is generated by taking adjacent pairs from the previous row and summing them, with `1` anchored at both ends.

## C++ Implementation
```cpp
vector<vector<int>> generate(int numRows) {
    vector<vector<int>> ans(numRows);

    for (int i = 0; i < numRows; i++) {

        ans[i].resize(i + 1);  
        ans[i][0] = ans[i][i] = 1;  // first + last = 1

        for (int j = 1; j < i; j++) {
            ans[i][j] = ans[i-1][j-1] + ans[i-1][j];
        }
    }

    return ans;
}
