```cpp
vector<vector<int>> fourSum(vector<int>& nums, int target) {
    vector<vector<int>> ans;          // stores all unique quadruplets
    int n = nums.size();

    // if fewer than 4 elements, no quadruplet possible
    if (n < 4) return ans;

    // sort array to enable two-pointer technique and duplicate handling
    sort(nums.begin(), nums.end());

    // first fixed element
    for (int i = 0; i < n - 3; i++) {

        // skip duplicate values for i to avoid repeated quadruplets
        if (i > 0 && nums[i] == nums[i - 1]) continue;

        // second fixed element
        for (int j = i + 1; j < n - 2; j++) {

            // skip duplicate values for j
            if (j > i + 1 && nums[j] == nums[j - 1]) continue;

            // minimum possible sum with current i and j
            // (smallest remaining two numbers)
            long long minSum =
                (long long)nums[i] + nums[j] + nums[j + 1] + nums[j + 2];

            // if even the smallest sum is greater than target,
            // no valid quadruplet further for this j
            if (minSum > target) break;

            // maximum possible sum with current i and j
            // (largest remaining two numbers)
            long long maxSum =
                (long long)nums[i] + nums[j] + nums[n - 1] + nums[n - 2];

            // if even the largest sum is smaller than target,
            // move to next j
            if (maxSum < target) continue;

            // two pointers for remaining two elements
            int p = j + 1;
            int q = n - 1;

            // standard two-pointer scan
            while (p < q) {

                // compute sum using long long to avoid overflow
                long long sum =
                    (long long)nums[i] + nums[j] + nums[p] + nums[q];

                // if sum is too small, move left pointer forward
                if (sum < target) {
                    p++;
                }
                // if sum is too large, move right pointer backward
                else if (sum > target) {
                    q--;
                }
                // valid quadruplet found
                else {
                    ans.push_back({nums[i], nums[j], nums[p], nums[q]});

                    // move both pointers after recording answer
                    p++;
                    q--;

                    // skip duplicate values for p
                    while (p < q && nums[p] == nums[p - 1]) p++;

                    // skip duplicate values for q
                    while (p < q && nums[q] == nums[q + 1]) q--;
                }
            }
        }
    }

    // return all unique quadruplets
    return ans;
}
