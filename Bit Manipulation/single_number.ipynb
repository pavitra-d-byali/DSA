# Single Number — XOR Logic (O(n), O(1))

## XOR Properties (Why This Works)
- `a ^ a = 0`  
- `0 ^ a = a`  
- XOR cancels out duplicates because pairs annihilate each other.
- If every element appears twice except one, XORing all of them leaves the unique element.

## Algorithm
1. Initialize `result = 0`
2. XOR every number into `result`
3. All duplicates vanish → the remaining value is the single non-duplicate number

## C++ Implementation
```cpp
class Solution {
public:
    int singleNumber(vector<int>& nums) {
        int result = 0;
        for (int n : nums) {
            result ^= n;   // XOR cancels duplicates
        }
        return result;
    }
};
