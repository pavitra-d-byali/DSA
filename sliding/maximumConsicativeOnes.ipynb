#include <vector>
#include <algorithm>
using namespace std;

class Solution {
public:
    int longestOnes(vector<int>& nums, int k) {
        int zeroCount = 0;   // number of zeros in current window
        int left = 0;        // left boundary of window
        int maxOnes = 0;     // maximum window size found

        for (int right = 0; right < nums.size(); right++) {

            // expand window by including nums[right]
            if (nums[right] == 0)
                zeroCount++;

            // shrink window until condition is satisfied
            while (zeroCount > k) {
                if (nums[left] == 0)
                    zeroCount--;
                left++;
            }

            // update maximum valid window length
            maxOnes = max(maxOnes, right - left + 1);
        }

        return maxOnes;
    }
};
