# Best Time to Buy and Sell Stock — Short Notes

## 1. Core Idea
Track two running values while scanning the array:

- **min_price** → lowest price seen so far (best day to buy)  
- **max_profit** → best profit found so far  

At each price:
profit_today = prices[i] - min_price

max_profit = max(max_profit, profit_today)

min_price = min(min_price, prices[i])

## 2. Why It Works
You only need:
- the lowest purchase price before today  
- the highest profit achievable if sold today  

This avoids checking all pairs. A single left-to-right scan respects the buy-before-sell rule and guarantees the optimal solution.

## 3. Complexity
- **Time:** O(n)  
- **Space:** O(1)  
This is already asymptotically optimal.

## 4. One-Line Summary
Track the lowest price so far and update the maximum profit each day.

## C++ Implementation
```cpp
class Solution {
public:
    int maxProfit(vector<int>& prices) {
        int min_price = prices[0];
        int max_profit = 0;

        for (int i = 1; i < prices.size(); i++) {
            int profit_today = prices[i] - min_price;
            max_profit = max(max_profit, profit_today);

            min_price = min(min_price, prices[i]);
        }

        return max_profit;
    }
};
